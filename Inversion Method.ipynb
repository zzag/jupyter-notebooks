```python
%matplotlib inline

import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

## Uniform distribution

```python
X_i = np.random.rand(1000)
Y_i = np.random.rand(1000)

plt.scatter(X_i, Y_i, marker=".")
```

## Sample Polar Coordinates

```python
N = 1000
r = np.random.rand(N)
theta = 2 * math.pi * np.random.rand(N)

plt.scatter(r * np.cos(theta), r * np.sin(theta), marker=".''")
```

Note that the samples are clumped at (0, 0).

## Sample Polar Coordinates (Uniform)

```python
N = 1000
r = np.sqrt(np.random.rand(1000))
theta = 2 * math.pi * np.random.rand(N)

plt.scatter(r * np.cos(theta), r * np.sin(theta), marker=".")
```

```python
X_i = np.arange(100)
Y_i = np.random.rand(100)
Y_ii = np.cos(Y_i)

test1 = np.count_nonzero(np.sort(Y_i) < Y_i[30])
print(test1)
test2 = np.count_nonzero(np.sort(Y_ii) > Y_ii[30])
print(test2)
```

```python
N = 100
X_i = np.arange(N)
la = 0.4
Y_i = -np.log(np.random.rand(100)) / la

plt.bar(X_i, Y_i)
```

```python
plt.bar(X_i, np.sort(Y_i)[::-1])
```
